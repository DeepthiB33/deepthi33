```python
def PrimeFactor(x):
    prime_fact = []
    d = 2

    while d <= x:
        if x % d == 0:
            prime_fact.append(d)
            x = x/d
        else:
            d += 1
    return prime_fact

print(PrimeFactor(34))
```
